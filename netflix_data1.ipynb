```python
#Process list of dicts and create formatted string repstn of genres
def make_genresList(x):
    gen = []
    st = " "
    for i in x:
        if i.get('name') == 'Science Fiction': #Renaming the "Science Fiction" to "Sci-Fi" to make it the name shorter
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen == []:
        return np.NaN
    else:
        return (st.join(gen))
```

```python
data2['genres']=data2['genres'].map(lambda x: make_genresList(x))
data2
```

```python
def make_castList(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] :
        return np.NaN
    else:
        return (casts[0])
```

```python
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn import naive_bayes
from sklearn.metrics import roc_auc_score,accuracy_score
import pickle
```

```python
nltk.download("stopwords")
```

```python
dataset = pd.read_csv("/Users/sathwikshroff/NetflixRecSystem/reviews.txt",sep = '\t', names =['Reviews','Comments'])
dataset
```

```python
stopset=set(stopwords.words('english'))
vectorizer=TfidfVectorizer(use_idf=True,lowercase=True,strip_accents='ascii',stop_words=stopset)
x=vectorizer.fit_transform(dataset.Comments)
y=dataset.Reviews
pickle.dump(vectorizer,open('transform.pkl','wb'))
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=50)
```

```python
clf=naive_bayes.MultinomialNB()
clf.fit(xtrain,ytrain)
```

```python
accuracy_score(ytest,clf.predict(xtest))
```

```python
pickle.dump(clf,open('model.pkl','wb'))
```
